```python
# import necessay library
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```


Dataset
The dataset I chose is the affairs dataset that comes with Statsmodels. It was derived from a survey of women in 1974 by Redbook magazine, in which married women were asked about their participation in extramarital affairs. More information about the study is available in a 1978 paper from the Journal of Political Economy.

```python
# import dataset:
dta=sm.datasets.fair.load_pandas().data
```

```python
dta.head()
```

```python
#add affair column
#where, 1=represents having affairs, and 0= represents not have affairs:
dta["affair"]=(dta.affairs>0).astype(int)
```

```python
dta.head()
```

```python
# preparing data for logistic Regression:
from patsy import dmatrices

y,X=dmatrices("affair ~ rate_marriage + age+ yrs_married + children + \
              religious + educ + C(occupation) + C(occupation_husb)",
             dta,return_type="dataframe")
X.columns
```

```python
#rename the columns
X=X.rename(columns={'C(occupation)[T.2.0]':'occ_2','C(occupation)[T.3.0]':'occ_3',
                    'C(occupation)[T.4.0]':'occ_4',
             'C(occupation)[T.5.0]':'occ_5','C(occupation)[T.6.0]':'occ_6',
             'C(occupation_husb)[T.2.0]':'occ_husb_2','C(occupation_husb)[T.3.0]':'occ_husb_3',
             'C(occupation_husb)[T.4.0]':'occ_husb_4','C(occupation_husb)[T.5.0]':'occ_husb_5',
             'C(occupation_husb)[T.6.0]':'occ_husb_6'})
```

```python
X
```

```python
# Flatten y into a 1D array
y=np.ravel(y)
#Return a contiguous flattened array.
y
```

```python
# Splitting the daaset into the Training set and Test set# Splitt 
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# Feature Scaling to maintain all the feature values in specific range
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test= sc.transform(X_test)  
```

```python
#Create the Moder
#Create teh model
from sklearn.linear_model import LogisticRegression
lgRegressor=LogisticRegression()

lgRegressor.fit(X_train,y_train)
```

```python
#predict the Moderl
y_predict=lgRegressor.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import classification_report
cReport=classification_report(y_test,y_predict)
print(cReport)
                    
```

```python
# generate evaluation metrics
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_predict))
print(metrics.roc_auc_score(y_test, y_predict))
```

```python
#Model Evaluation Using Cross-Validation
# evaluate the model using 10-fold cross-validation
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=10)
print(scores)
print(scores.mean())
```
